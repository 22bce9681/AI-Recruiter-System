```python
import json

candidates = []

with open('/content/india_runs/[PUB] India_runs_data_and_ai_challenge/India_runs_data_and_ai_challenge/candidates.jsonl', 'r') as f:
    for line in f:
        candidates.append(json.loads(line))

print("Total Candidates:", len(candidates))
print("\nFirst Candidate:\n")
print(candidates[0])
```

```python
from docx import Document

doc = Document(
    '/content/india_runs/[PUB] India_runs_data_and_ai_challenge/India_runs_data_and_ai_challenge/job_description.docx'
)

job_text = ""

for para in doc.paragraphs:
    job_text += para.text + " "

print(job_text)
```

```python
candidate_texts = []

for candidate in candidates:
    profile = candidate.get('profile', {})

    text = ""

    # Skills
    skills = profile.get('skills', [])
    text += " ".join(skills) + " "

    # Experience
    for exp in profile.get('experience', []):
        text += exp.get('title', '') + " "
        text += exp.get('company_name', '') + " "
        text += exp.get('description', '') + " "

    candidate_texts.append(text)

print("Total Candidate Texts:", len(candidate_texts))
print("\nSample Candidate Text:\n")
print(candidate_texts[0][:1000])
```

```python
from sentence_transformers import SentenceTransformer

# Load SBERT model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Job embedding
job_embedding = model.encode(job_text)

# Use only first 1000 candidates for testing
sample_texts = candidate_texts[:1000]
sample_ids = [c['candidate_id'] for c in candidates[:1000]]

# Candidate embeddings
candidate_embeddings = model.encode(
    sample_texts,
    batch_size=32,
    show_progress_bar=True
)

print("Job embedding size:", len(job_embedding))
print("Candidate embeddings shape:", candidate_embeddings.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

# Calculate similarity scores
scores = cosine_similarity(
    [job_embedding],
    candidate_embeddings
)[0]

# Create ranking dataframe
results = pd.DataFrame({
    'candidate_id': sample_ids,
    'score': scores
})

# Sort by score
results = results.sort_values(
    by='score',
    ascending=False
)

# Show top 10 candidates
results.head(10)
```

```python
top_candidates = results.head(3)

for _, row in top_candidates.iterrows():

    print("=" * 60)
    print("Candidate ID:", row['candidate_id'])
    print("Match Score:", round(row['score'] * 100, 2), "%")
    print("Why Selected:")
    print("✓ High semantic similarity with job description")
    print()
```

```python
for _, row in results.head(5).iterrows():

    candidate = next(
        c for c in candidates
        if c['candidate_id'] == row['candidate_id']
    )

    profile = candidate.get('profile', {})

    print("=" * 60)
    print("Candidate ID:", row['candidate_id'])
    print("Match Score:", round(row['score'] * 100, 2), "%")

    print("\nSkills:")
    skills = profile.get('skills', [])

    if len(skills) > 0:
        print(", ".join(skills[:10]))
    else:
        print("No skills available")

    print("\nExperience:")

    experiences = profile.get('experience', [])

    if len(experiences) > 0:
        for exp in experiences[:2]:
            print("-", exp.get('title', 'N/A'),
                  "|", exp.get('company_name', 'N/A'))
    else:
        print("No experience available")

    print("\nWhy Selected:")
    print("✓ High semantic similarity")
    print("✓ Relevant profile found")
    print()
```

```python
top_candidates = results.head(10).copy()

def recommendation(score):

    score = score * 100

    if score >= 70:
        return "Strongly Recommend"

    elif score >= 40:
        return "Recommend"

    else:
        return "Consider"

top_candidates["Recommendation"] = (
    top_candidates["score"].apply(recommendation)
)

top_candidates["Match %"] = (
    top_candidates["score"] * 100
).round(2)

print(
    top_candidates[
        ["candidate_id", "Match %", "Recommendation"]
    ]
)
```

```python
import matplotlib.pyplot as plt

top5 = results.head(5)

plt.figure(figsize=(8,5))

plt.bar(
    top5["candidate_id"],
    top5["score"] * 100
)

plt.xlabel("Candidate ID")
plt.ylabel("Match Percentage")
plt.title("Top 5 Candidate Matches")

plt.xticks(rotation=45)

for i, score in enumerate(top5["score"] * 100):
    plt.text(i, score + 0.1, f"{score:.1f}%")

plt.show()
```

```python
final_results = top_candidates[
    ["candidate_id", "Match %", "Recommendation"]
]

final_results.to_csv(
    "shortlisted_candidates.csv",
    index=False
)

print("CSV file saved successfully!")
```
